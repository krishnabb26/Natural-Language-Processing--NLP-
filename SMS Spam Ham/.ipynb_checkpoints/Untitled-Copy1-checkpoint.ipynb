```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

import warnings
warnings.filterwarnings('ignore')
```

```python
#Data Cleaning and PreProcessing

ps = PorterStemmer()
wnl = WordNetLemmatizer()


corpus = []

for i in range(0, len(sms)):
    review = re.sub('[^a-zA-Z]',' ',sms['messages'][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
sms.loc[sms.labels == 'ham', 'labels'] = 0
sms.loc[sms.labels == 'spam','labels'] = 1

x = corpus
y = sms['labels']
y = y.astype(int)
```

```python
y.shape
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
cv = CountVectorizer(max_features=4900)
x_train = cv.fit_transform(xtrain)
x_test = cv.transform(xtest)
```

```python
ytrain.value_counts()
```

```python
ytest.value_counts()
```

```python
ytrain.value_counts()
```

```python
ytest.value_counts()
```

```python
#Training the Model

from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB()

spam_detect_model.fit(x_train, ytrain) 
ypred = spam_detect_model.predict(x_test)
train = spam_detect_model.score(x_train, ytrain)
test = spam_detect_model.score(x_test, ytest)

print(f'TRAIN ACCURACY: {train}\nTEST ACCURACY: {test}\n\n{classification_report(ytest, ypred)}')
```

```python
ypred
```

```python
from sklearn.metrics import confusion_matrix

confusion_m =confusion_matrix(ytest, ypred)

sns.heatmap(confusion_m, cmap = 'bone',annot = True, fmt = '.1f',
            xticklabels = ['Ham', 'Spam'], yticklabels = ['Ham', 'Spam'])
plt.xlabel('Predicted Output')
plt.ylabel('Actual Output')
plt.show();
```
